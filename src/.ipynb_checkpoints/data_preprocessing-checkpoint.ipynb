```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_json('../dataset/Software_5.json', lines=True)
```

```python
dataset
```

```python
dataset.iloc[0,:]
```

```python
unused_columns = [
    "verified",
    "reviewTime",
    "reviewerID",
    "asin",
    "style",
    "reviewerName",
    "summary",
    "unixReviewTime",
    "vote",
    "image"
]
dataset.drop(columns=unused_columns, inplace=True)
```

```python
dataset
```

```python
dataset[dataset["overall"] == 1]
```

```python
dataset = dataset[[col if col != "overall" and col != "reviewText" else "overall" if col == "reviewText" else "reviewText" for col in dataset.columns]]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset[dataset["reviewText"].isna()]
```

```python
dataset.dropna(inplace=True)
```

```python
dataset[dataset["reviewText"].isna()]
```

```python
from collections import Counter
sentiment_distr = Counter(dataset["overall"])
print(sentiment_distr)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(dataset, test_size = 0.33, random_state=42)
```

```python
train
```

```python
test
```

```python
train_sentiment_distr = Counter(train["overall"])
print(train_sentiment_distr)
```

```python
test_sentiment_distr = Counter(test["overall"])
print(test_sentiment_distr)
```

```python
train.to_csv('../input/train.tsv', sep='\t', index=False)
```

```python
test.to_csv('../input/test.tsv', sep='\t', index=False)
```
